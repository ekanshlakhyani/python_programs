## Some Helper function for our main function

```python
def xrecursion(i, n, s_space, m_space):
  '''
  Returns None, consuming i which is the variable(natural number) for help in 
  recursion, n which is the number of x in a row or a column, s_space which
  helps in number of spaces before and after the first and the second x in a 
  row, and m_space which helps with the number of middle spaces in between
  the two x. 
  
  xrecursion: Nat Nat Nat Int -> None
  Requires: n > 0
            i = n
            s_space = 0
            m_space = n - 2
  
  Effects: Prints the xbox for the given n.
  
  Examples:
        xrecursion(1, 1, 0, -1) => None
        and prints
        ###
        #x#
        ###
        
        xrecursion(5, 5, 0, 3) => None
        and prints
        #######
        #x   x#
        # x x #
        #  x  #
        # x x #
        #x   x#
        #######
        
        xrecursion(10, 10, 0, 8) => None
        and prints
        ############
        #x        x#
        # x      x #
        #  x    x  #
        #   x  x   #
        #    xx    #
        #    xx    #
        #   x  x   #
        #  x    x  #
        # x      x #
        #x        x#
        ############
  '''
  if i == 0:
    print('#' * (n + 2))
    return None
  elif math.ceil(n/2) < i:
    print ('#'+ s_space*' ' + 'x' + m_space*' '  + 'x' +  s_space*' ' +'#')
    return xrecursion(i-1, n, s_space+1, m_space-2)
  elif (n % 2) != 0:
    if math.ceil(n/2) == i:
      print ('#'+ s_space*' ' + 'x' +  s_space*' ' +'#')
      return xrecursion(i-1, n, s_space-1, m_space+2)
    else:
      print ('#'+ s_space*' ' + 'x' + m_space*' '  + 'x' +  s_space*' ' +'#')
      return xrecursion(i-1, n, s_space-1, m_space+2)
  else:
    print ('#'+(s_space-1)*' '+'x'+(m_space+2)*' '+'x'+(s_space-1)*' '+'#')
    return xrecursion(i-1, n, s_space-1, m_space+2)
```

## Our main function

```python
def xbox(n):
  '''
  Returns None, consuming n, whic is the number of rows and columns of x in the
  printed x box.
  
  xrecursion: Nat -> None
  Requires: n > 0
            
  Effects: Prints the xbox for the given n.
  
  Examples:
        xbox(1) => None
        and prints
        ###
        #x#
        ###
        
        xbox(5) => None
        and prints
        #######
        #x   x#
        # x x #
        #  x  #
        # x x #
        #x   x#
        #######
        
        xbox(10) => None
        and prints
        ############
        #x        x#
        # x      x #
        #  x    x  #
        #   x  x   #
        #    xx    #
        #    xx    #
        #   x  x   #
        #  x    x  #
        # x      x #
        #x        x#
        ############
  '''
  print('#' * (n + 2))
  return xrecursion(n, n, 0, n-2)
```
